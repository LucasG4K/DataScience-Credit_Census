#### **APPLICATION OF DECISION TREE(DT) ALGORITHM**

---
- **BASE CREDIT**

```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import pickle
```

```python
with open('data/credit.pkl', 'rb') as file:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(file)
```

```python
x_credit_test.shape, y_credit_test.shape
```

```python
x_credit_training.shape, y_credit_training.shape
```

```python
dt_credit = DecisionTreeClassifier(criterion='entropy')
dt_credit.fit(x_credit_training, y_credit_training)
```

```python
prediction = dt_credit.predict(x_credit_test)
```

```python
prediction - y_credit_test
```

```python
accuracy_score(y_credit_test, prediction) # 0.982
```

```python
sns.heatmap(confusion_matrix(y_credit_test, prediction), annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
print(classification_report(y_credit_test, prediction))
```

```python
plt.figure(figsize=(15,15))
plot_tree(
    dt_credit,
    feature_names = ['income', 'age', 'loan'],
    class_names=[str(value) for value in dt_credit.classes_],
    filled=True
);
```
