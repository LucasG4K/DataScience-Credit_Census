#### **APPLICATION OF LOGISTIC REGRESSION(LR) ALGORITHM**

---
- **BASE CREDIT**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import pickle
```

```python
with open('data/credit.pkl', 'rb') as file:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(file)
```

```python
x_credit_test.shape, y_credit_test.shape
```

```python
x_credit_training.shape, y_credit_training.shape
```

```python
lr_credit = LogisticRegression(random_state=1)
lr_credit.fit(x_credit_training, y_credit_training)
```

```python
prediction = lr_credit.predict(x_credit_test)
```

```python
prediction - y_credit_test
```

```python
accuracy_score(y_credit_test, prediction)
```

```python
sns.heatmap(confusion_matrix(y_credit_test, prediction), annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
print(classification_report(y_credit_test, prediction))
```

```python
with open('data/census.pkl', 'rb') as file:
    x_census_training, y_census_training, x_census_test, y_census_test = pickle.load(file)
```

```python
sns.heatmap(confusion_matrix(y_census_test, prediction2), annot=True, fmt='d', cmap='YlGnBu', xticklabels=['<=50K', '>50K'], yticklabels=['<=50K', '>50K'])
plt.yticks(rotation=0)
plt.show()
```

```python
print(classification_report(y_census_test, prediction2))
```
