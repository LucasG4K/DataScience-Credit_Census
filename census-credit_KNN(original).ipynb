#### **APPLICATION OF K-NEAREST NEIGHBOR(KNN) ALGORITHM**

---
- **BASE CREDIT**

```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import pickle
```

```python
with open('data/credit.pkl', 'rb') as file:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(file)
```

```python
x_credit_test.shape, y_credit_test.shape
```

```python
x_credit_training.shape, y_credit_training.shape
```

```python
knn_credit = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_credit.fit(x_credit_training, y_credit_training)
```

```python
prediction = knn_credit.predict(x_credit_test)
```

```python
prediction - y_credit_test
```

```python
accuracy_score(y_credit_test, prediction) # 0.986
```

```python
sns.heatmap(confusion_matrix(y_credit_test, prediction), annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
print(classification_report(y_credit_test, prediction))
```

- **BASE CENSUS**

```python
with open('data/census.pkl', 'rb') as file:
    x_census_training, y_census_training, x_census_test, y_census_test = pickle.load(file)
```

```python
x_census_test.shape, y_census_test.shape
```

```python
x_census_training.shape, y_census_training.shape
```

```python
knn_census = KNeighborsClassifier(n_neighbors=10)
knn_census.fit(x_census_training, y_census_training)
```

```python
prediction2 = knn_census.predict(x_census_test)
prediction2
```

```python
accuracy_score(y_census_test, prediction2) # 0.834296961495879
```

```python
sns.heatmap(confusion_matrix(y_census_test, prediction2), annot=True, fmt='d', cmap='YlGnBu', xticklabels=['<=50K', '>50K'], yticklabels=['<=50K', '>50K'])
plt.yticks(rotation=0)
plt.show()
```

```python
print(classification_report(y_census_test, prediction2))
```
